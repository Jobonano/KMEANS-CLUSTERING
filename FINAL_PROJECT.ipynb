```python
from sklearn.impute import SimpleImputer
```

```python
imputer=SimpleImputer(strategy='median')
df_imputed = pd.DataFrame(imputer.fit_transform(df.drop('CUST_ID', axis=1)))
```

```python
df_imputed.isnull().sum()
```

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
kmeans = KMeans(n_clusters=2, init='k-means++', max_iter=300, tol=1e-4, random_state=101, algorithm='elkan')
```

```python
kmeans.fit(df_imputed)
```

```python
prediction= kmeans.predict(df_imputed)
```

```python
prediction
```

```python
labels = kmeans.labels_
```

```python
from sklearn.metrics import silhouette_score
```

```python
#Calculating the silhouette score
silhouette_avg = silhouette_score(df_imputed, labels)
print(f"Silhouette Score: {silhouette_avg}")
```

```python
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(df_imputed)
```

```python
# Plotting the clusters in 2D
plt.figure(figsize=(8, 6))
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=labels, cmap='viridis', marker='.')
plt.title('KMeans Clustering (2 Clusters)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster')
plt.show()
```
